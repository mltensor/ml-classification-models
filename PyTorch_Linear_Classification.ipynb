### 1. Data (preparing and loading)
Data can be anything... in ML

Machine learning is a game of two parts:
1. Get data into a numerical representation.
2. Build a model to learn patterns in that numerical representation.

To showcase this, let's create some *known* data using linear regression formula.

We will use linear regression formula to make a straight line with known *parameters*.

```python
# Cretae known parameters
weight = 0.7
bias = 0.3

# Create
start = 0
end = 1
step = 0.02
X = torch.arange(start, end, step)
y = weight * X + bias

X[:10], y[:10], X.dtype
```

```python
len(X), len(y)
```

### Splitting data into training and test sets

Let's create a training and testing set of our data

```python
# Create a train/test split
train_split = int(0.8 * len(X))
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]
len(X_train), len(y_train), len(X_test), len(y_test)
```

### Visualisation of our data

This is where data explorer's motto comes in

```python
def plot_predictions(train_data=X_train,
                     train_label=y_train,
                     test_data=X_test,
                     test_label=y_test,
                     predictions=None):
  """
  Plots training data, test data and compares predictions.
  """
  plt.figure(figsize=(10,7))

  # Plot taining data in blue
  plt.scatter(train_data, train_label, c="b", s=4, label="Training Data") # s is the size of the dots used by matplotlib
  plt.scatter(test_data, test_label, c="g", s=4, label="Testing Data")

  # Are there predictions?
  if predictions is not None:
    plt.scatter(test_data, predictions, c="r", s=4, label="Predictions")

  # show the legend
  plt.legend(prop={"size": 14});
```

```python
plot_predictions()
```

### 2. Build model
Our first PyTorch model!

For updating the weights we have to main algorithms:
1. Gradient Descent
2. Back Propagation

```python
# Create a linear regression model class
from torch import nn

class LinearRegressionModel(nn.Module): # <- almost everthing in PyTrch inherits nn.Module
  def __init__(self):
    super().__init__()
    self.weights = nn.Parameter(torch.randn(1, # 1 here represent the variance of the random number generated
                                            requires_grad=True, # it will allow PyTorch to keep track of gradient used
                                            dtype=torch.float))
    self.bias = nn.Parameter(torch.randn(1,
                                         requires_grad=True,
                                         dtype=torch.float))

  # Forward to define the computation in the model
  def forward(self, x:torch.Tensor) -> torch.Tensor: # <- "x" is the input data
    return self.weights * x + self.bias # this is the line regression formula
  # any subclass of nn.Module needs to override forward()
  # (this defines the forward computation of the model)
```

### PyTorch model building essentials

* torch.nn - contains all the buildings for computational graphs
* torch.nn.Parameter - what parameters should our model try and learn, often a PyTorch layer from torch.nn will set these for us
* torch.nn.Module - The base class for all neural network modules, if you subclass it, you should overwrite forward()
* torch.optim - It is optimizer
* def forward() - All nn.Module subclasses require you to overwrite forward, this method define what happens in the forward computation

### 3. Train Model

From poor representation of the data to a better representation of the data

One way to measure how poor or how wrong our models are is to use loss function

* Note: Loss functions is almost same as cost functions

Things we need to train:
* Loss Function - we will use `L1Loss` functions which is basically the mean of absolute error.
* Optimizer - we will use `SGD`, which is stochastic gradient descent... well you know

```python
model_0.state_dict(), list(model_0.parameters())
```

```python
# Setup a loss function
loss_fn = nn.L1Loss() # we are just defining the variables here

# Setup an optimizer
optimizer = torch.optim.SGD(params=model_0.parameters(),
                            lr=0.01) # learning rate
```

### Building a training loop (and a testing loop) in PyTorch

Things we need in a training loop:
0. Loop through the data
1. forward pass (this involves data moving through our model functions) -  also called forward propagations
2. Calculate the loss (compared forward pass predictions to ground truth level)
3. Optimizer zero depth
4. Loss backward - move backward through network to calculate the gradients of each of the model
5. Optimizer step - use the optimizer to adjust out model's parameters to try and improve the loss

```python
with torch.inference_mode():
  list(model_0.parameters())
```

```python
list(model_0.parameters())
```

```python
%%time
torch.manual_seed(42)
# An epoch is one loop through the data...
epochs = 1000
epoch_count=[]
train_loss_values=[]
test_loss_values=[]

# 0. Looping through the data
for epoch in range(epochs):
  # set the model to training mode
  model_0.train() # train mode in PyTorch sets all parameters that require gradients to to require gradients

  # 1. Forward pass
  y_pred = model_0(X_train)

  # 2. Calculate the loss
  loss = loss_fn(y_pred, y_train)
  # print(f"Loss: {loss}")

  # 3. Optimizer zero grad
  optimizer.zero_grad() # set the gradient of all optimizer to zero... to start with fresh gradient for each trainig set

  # 4. Perform back propagation on the loss with respect to the parametes of the model
  loss.backward()

  # 5. Step the optimizer (perform gradient descent)
  optimizer.step() # by default how the optimizer changes will accumulate through the loop.... we have ot zero them above in step 3

  # Testing
  model_0.eval() # turn off different settings in the model not needed for evaluation/testing
  with torch.inference_mode(): # turns off the gradietn tracking & couple of more things behind the scene
    # 1. Forward pass
    test_pred = model_0(X_test)

    # 2. Calculate the test loss
    test_loss = loss_fn(test_pred, y_test)
  # Print what's happening
  if epoch % 100==0:
    epoch_count.append(epoch)
    train_loss_values.append(loss)
    test_loss_values.append(test_loss)
    print(f"Epoch: {epoch} | Loss: {loss} | Test loss: {test_loss}")
    # Pritn out the model state_dixt()
    print(model_0.state_dict())
```

```python
import numpy as np
np.array(torch.tensor(test_loss_values).cpu().numpy()), test_loss_values
```

```python
# Plot the loss curve
plt.plot(epoch_count, np.array(torch.tensor(train_loss_values).cpu().numpy()), label="Train Loss")
plt.plot(epoch_count, np.array(torch.tensor(test_loss_values).cpu().numpy()), label="Test Loss")
plt.title("Training and test loss curve")
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.legend()
```

### Saving a model in PyTorch
There are three main methods.
* `torch.save()` - allows you to save a PyTorch object in Python's pickle format.
* `torch.load()` - allows you to load a saved PyTorch object.
* `torch.nn.MOdule.load_state_dict()` - this allows to load a model's saved state dictionary.

```python
model_0.state_dict()
```

* Find the pros and cons of saving and loading the entire model vs saving in `state_dict`

```python
# saving a model
from pathlib import Path

# 1. Create models directory
MODEL_PATH = Path("models")
MODEL_PATH.mkdir(parents=True, exist_ok=True) # exist_ok will throw no error even if the file already exist

# 2. Create model save path
MODEL_NAME = "Learning_PyTorch.pth" # can be used .pt as well
MODEL_SAVE_PATH = MODEL_PATH / MODEL_NAME

MODEL_SAVE_PATH

# 3. Save the model using state_dict
print(f"Saving model to: {MODEL_SAVE_PATH}")
torch.save(obj=model_0.state_dict(),
           f=MODEL_SAVE_PATH)
```

Since we saved our model's `state_dict()` rather the entire model

```python
model_0.state_dict()
```
